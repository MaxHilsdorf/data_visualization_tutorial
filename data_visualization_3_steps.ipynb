# Data Visualization in Three Steps

In this notebook, I'll share the code for my Medium Article: ["From Data Analyst to Data Storyteller in Three Steps"](https://towardsdatascience.com/from-data-analyst-to-data-storyteller-in-3-steps-54b56815b567)

## Data Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
music_pref = pd.read_csv("DIR/preferred_music.csv",
                        delimiter = ";")
music_pref.head()
```

```python
music_pref_2019 = music_pref[music_pref["year"] == 2019]
```

```python
genres = music_pref.columns[2:]
```

## Visualizations

```python
# Set plot style template
plt.style.use("default")

# Define colors (Tableaus Colorblind 10)
cb_colors = [(0/255, 107/255, 164/255),
            (1/255, 128/255, 14/255),
            (171/255, 171/255, 171/255),
            (89/255, 89/255, 89/255),
            (95/255, 158/255, 205/255),
            (200/255, 82/255, 0/255),
            (137/255, 137/255, 137/255),
            (162/255, 200/255, 236/255),
            (255/255, 188/255, 121/255),
            (207/255, 207/255, 207/255)]

title_font= {"family" : "Cambria",
             "size" : 15,
             "color" : "black",
             "weight" : "roman"}
```

### Basic Plot

```python
fig, ax = plt.subplots(figsize =(10,6))

# Loop through the genres
for i, g in enumerate(genres):
    
    # Define x and y
    x = np.arange(7)
    y = music_pref_2019[g]

    # Plot dots
    #ax.scatter(x, y, label = g)
    
    # Plot lines
    ax.plot(x, y, label = g)
    
    # X axis
    ax.set_xticks(x)
    ax.set_xticklabels(["14-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70+"])
    
ax.legend()


#plt.grid(axis = "x", alpha = 0.2)
plt.savefig("figures/preferences_by_age_raw.png", dpi = 200)


plt.show()
```

### Step 1

```python
fig, ax = plt.subplots(figsize =(10,6))

# Loop through the genres
for i, g in enumerate(genres):
    
    # Define x and y
    x = np.arange(7)
    y = music_pref_2019[g]

    # Plot dots
    ax.scatter(x, y, c = cb_colors[i])
    
    # Plot lines
    ax.plot(x, y, label = g, c = cb_colors[i])
    
    # X axis
    ax.set_xticks(x)
    ax.set_xticklabels(["14-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70+ years"])
    
    # Y axis
    ax.set_ylim(0,1)
    ax.set_yticks([0,0.25,0.5,0.75,1])
    ax.set_yticklabels([0, 25, 50, 75, "100%"])


    # Overall
    ax.set_title("Popularity of Different Genres by Age Groups")
    
    ax.tick_params(bottom = False, top = False,
                  left = False, right = False)
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    # Annotations etc.
    
ax.text(0,.875, "Rock/Pop", rotation = -1.5)
ax.text(0,.144, "Classical", rotation = 16)
ax.text(0,.395, "Hardrock/Metal", rotation = 4.5)
ax.text(0,.679, "Hip Hop/Rap", rotation = -31)
ax.text(0,.592, "Techno/House", rotation = -15)
    

plt.savefig("figures/preferences_by_age_reduced.png", dpi = 200)


plt.show()
```

### Step 2

```python
fig, ax = plt.subplots(figsize =(10,6))

# Loop through the genres
for i, g in enumerate(genres):
    
    # Define x and y
    x = np.arange(7)
    y = music_pref_2019[g]

    # Plot dots
    ax.scatter(x, y, c = cb_colors[i])
    
    # Plot lines
    ax.plot(x, y, label = g, c = cb_colors[i])
    
    # X axis
    ax.set_xticks(x)
    ax.set_xticklabels(["14-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70+ years"])
    
    # Y axis
    ax.set_ylim(0,1)
    ax.set_yticks([0,0.25,0.5,0.75,1])
    ax.set_yticklabels([0, 25, 50, 75, "100%"])


    # Overall
    ax.set_title("Popularity of Different Genres by Age Groups")
    
    ax.tick_params(bottom = False, top = False,
                  left = False, right = False)
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    # Annotations etc.
    
ax.text(0,.875, "Rock/Pop", rotation = -1.5)
ax.text(0,.144, "Classical", rotation = 16)
ax.text(0,.395, "Hardrock/Metal", rotation = 4.5)
ax.text(0,.679, "Hip Hop/Rap", rotation = -31)
ax.text(0,.592, "Techno/House", rotation = -15)

ax.plot(0)
    

plt.savefig("figures/preferences_by_age_emphasized.png", dpi = 200)


plt.show()
```

### Step 3

```python
fig, ax = plt.subplots(figsize =(10,6))

# Setup a dict to map the column names to their actual string format
mapping_dict = {"rock_pop" : "Rock/Pop",
               "hip_hop" : "Hip Hop/Rap",
               "techno_house" : "Techno/House",
               "hardrock_heavy_metal" : "Hardrock/Metal",
               "classical" : "Classical"}

# Loop through the genres
for i, g in enumerate(genres):
    
    # Define x and y
    x = np.arange(7)
    y = music_pref_2019[g]

    # Plot dots
    ax.scatter(x, y, c = cb_colors[i], label = mapping_dict[g], marker = ".", linewidth = 3, alpha = 0.8)
    
    # Plot lines
    ax.plot(x,y, c = cb_colors[i], linestyle = "-", linewidth = 2.5, alpha = 0.7)
    
    # X axis
    ax.set_xticks(x)
    ax.set_xticklabels(["17", "25", "35", "45", "55", "65", "70+ years"])
    #ax.xaxis.grid(True)
    #ax.set_xlabel("Age (Years)")
    
    # Y axis
    ax.set_ylim(0,1)
    ax.set_yticks([0, 0.25, 0.5, 0.75, 1])
    ax.set_yticklabels([0, 25, 50, 75, "100%"])
    #ax.set_ylabel("People with Preference (%)")
    
    # Overall
    ax.set_title("Popularity of Different Genres by Age Groups", fontdict = title_font)
    
    ax.tick_params(bottom = False, top = False,
                  left = False, right = False)
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
# Annotations etc.
    
ax.text(0,.875, "Rock/Pop", rotation = -1.5, size = 11.5)
ax.text(0,.144, "Classical", rotation = 16, size = 11.5)
ax.text(0,.395, "Hardrock/Metal", rotation = 4.5, size = 11.5)
ax.text(0,.66, "Hip Hop/Rap", rotation = -31, size = 11.5)
ax.text(0,.581, "Techno/House", rotation = -15, size = 11.5)


#ax.grid(axis = "x", alpha = 0.2)
plt.savefig("figures/preferences_by_age.png", dpi = 200)

plt.show()
```
